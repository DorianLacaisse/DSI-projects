# Movies' rating prediction

The objective of that project is to understand the imdb rating and try to predict it in function of different features.  

We will first get the data, from the imdb website (title, Budget,..) and also from omdbapi which has an API and contains several data about each film.  
To get a sample reflecting any kind of movies and ranking, I will select around 1000 movie from 2013 to 2015.

After cleaning the data, I ll explorate and select the features that will be part of my predictor.

Then, as Amazon use RandomForest regressor, I ll make different tree decision regressor work against each other to then select the one who got the better result.

To end, we will compare the imdb rate predicted from the selected model to the actual one.

R2 score and mean squared error will be the metrics used to validate the model

```python
# Imports at the top
import json
import urllib
import pandas as pd
import numpy as np
import requests
import json
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#we will first create a function to retrieve the movie-code of 350 movies per years from imdb websites
def get_350_movies_year(year):
    entries=[]
    for page in range(1,8):
        response = requests.get('http://www.imdb.com/search/title?year='+year+','+year+'&title_type=feature&sort=moviemeter,asc&page='+str(page)+'&ref_=adv_nxt')
        html = response.text
        entries_page = re.findall("<a href.*?/title/(.*?)/", html) 
        entries_page=list(set(entries_page))
        entries=entries+entries_page
    return entries
```

```python
# let's create a list gather more than a thousand movie-code from 2013 to 2015

entries2015=get_350_movies_year('2015')
entries2014=get_350_movies_year('2014')
entries2013=get_350_movies_year('2013')
entries=entries2013+entries2014+entries2015
entries=[c for c in entries if len(c)==len(entries2015[0])]
```

```python
#this function will enable to retrieve information of a movie on the website omdbapi
def get_entry(entry):
    res = requests.get('http://www.omdbapi.com/?i='+entry)
    if res.status_code != 200:
        print entry, res.status_code
    else:
        print '.',
    try:
        j = json.loads(res.text)
    except ValueError:
        j = None
    return j
```

```python
# a budget might link to the success of the movie, the following function will retrieve it

def get_budget(entry):
    response = requests.get('http://www.imdb.com/title/'+entry)
    html = response.text
    try:
        budget_list = re.findall("Budget:</h4>[ ]*\$([^ ]*)", html)
        budget = int(budget_list[0].replace(',', ''))
        print '.',
        return budget
    except Exception as ex:
        print
        print ex, entry, response.status_code
        return None
```

### Data Cleaning

```python
#let's clan the data
df = df.replace('N/A', np.nan)
df.info()
```

```python
df.Released = pd.to_datetime(df.Released)
```

```python
def intminutes(x):
    try:
        y = x.replace('min', '').strip()
        return int(y)
    except Exception:
        return np.nan

df.Runtime = df.Runtime.apply(intminutes)
```

```python
df=df.loc[df.Runtime.isnull()==False,:]
df.reset_index(inplace=True,drop=True)
```

```python
df.Year = df.Year.astype(int)
```

```python
df.imdbRating = df.imdbRating.astype(float)
```

```python
def intvotes(x):
    try:
        y = x.replace(',', '').strip()
        return int(y)
    except Exception:
        return np.nan
df.imdbVotes = df.imdbVotes.apply(intvotes)
```

```python
df=df.loc[df.imdbVotes.isnull()==False,:]
df.reset_index(inplace=True,drop=True)
```

```python
pd.set_option("display.max_columns",200)
```

```python
#df = pd.merge(df, df3)
#df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=df.loc[df.Genre.isnull()==False,:]
df.reset_index(inplace=True,drop=True)
```

```python
# let's get the dummies for the genre of movie.
cv = CountVectorizer()
data = cv.fit_transform(df.Genre).todense()
columns = ['genre_'+c for c in cv.get_feature_names()]
genredf = pd.DataFrame(data, columns=columns)
genredf.head()
```

```python
# let's get the dummies for actors categories.
cv = CountVectorizer(token_pattern=u'(?u)\\w+\.?\\w?\.? \\w+')
data = cv.fit_transform(df.Actors).todense()
columns = ['actor: '+c for c in cv.get_feature_names()]
actorsdf = pd.DataFrame(data, columns=columns)

# 100 major actors:
top100_actors= actorsdf.apply(sum).sort_values(ascending=False)
top100_actors=top100_actors[0:100].index
actorsdf=actorsdf.loc[:,top100_actors.tolist()]
actorsdf.head()
```

```python
df=pd.concat([df,genredf],axis=1)
```

```python
df=pd.concat([df,actorsdf],axis=1)
```

```python
# let's get the dummies for the Directors category.
cv = CountVectorizer(token_pattern=u'(?u)\\w+\.?\\w?\.? \\w+')
data = cv.fit_transform(df.Director).todense()
columns = ['director: '+c for c in cv.get_feature_names()]
directordf = pd.DataFrame(data, columns=columns)
```

```python
# 100 major directors:
top100_directors= directordf.apply(sum).sort_values(ascending=False)
top100_directors=top100_directors[0:100].index
directordf=directordf.loc[:,top100_directors.tolist()]
directordf.head()
```

```python
df=pd.concat([df,directordf],axis=1)
```

```python
# let's delete some columns we won't work with so that to get a better visibility
todelete=["Actors",'Awards',"Poster","Writer","Genre","Director"]

for item in todelete:
    del df[item]

df.head()
```

```python
#deleting row with Metascore as NaN value
df=df.loc[df.Metascore.isnull()==False,:]
df.reset_index(inplace=True,drop=True)
```

```python
df.Metascore=df.Metascore.astype('int')
```

```python
#Let's scale all the column so that to get a number between 0 to 1.
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

Xt=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
Xt.head()
```

```python
# Before starting, let's have a look to the correlation between imdb rating and Metascore

plt.scatter(df["Metascore"],df["imdbRating"])
plt.xlim((0,100))
plt.xlabel("Metascore")
plt.ylabel("imdbRating")
plt.show()
```

It seem's to have a slight correlation between imdb rating and Metascore, it will probably be one of the most important feature for the DecisionTree.

```python
# let's create our train and test set:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Xt,y,test_size=0.3, random_state=42)
```

```python
# Let's now use the Random forest regressor to predict the imdb score.
# we'll first choose the best number of feature.

from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor, ExtraTreesRegressor, RandomForestRegressor
from sklearn.model_selection import cross_val_score, cross_val_predict

estimators=range(1,Xt.shape[1],10)

def get_mean_mse(model,n):
    ensemble=model(n_estimators=n)
    score=cross_val_score(ensemble,X_train,y_train, cv=5,scoring="neg_mean_squared_error")
    return np.mean(score)

Bagging=[]
Adaboost=[]
ExtraTree=[]
RandomForest=[]


for i in estimators:
    Bagging.append(get_mean_mse(BaggingRegressor,i))
    Adaboost.append(get_mean_mse(AdaBoostRegressor,i))
    ExtraTree.append(get_mean_mse(ExtraTreesRegressor,i))
    RandomForest.append(get_mean_mse(RandomForestRegressor,i))

print("min Bagging:"+ str(min(np.absolute(Bagging)))+ "for n-estimator:" + \
str(estimators[Bagging.index(-min(np.absolute(Bagging)))]))

print("min Adabost:"+ str(min(np.absolute(Adaboost)))+ "for n-estimator:" + \
str(estimators[Adaboost.index(-min(np.absolute(Adaboost)))]))

print("min ExtraTree:"+ str(min(np.absolute(ExtraTree)))+ "for n-estimator:" + \
str(estimators[ExtraTree.index(-min(np.absolute(ExtraTree)))]))

print("min RandomForest:"+ str(min(np.absolute(RandomForest)))+ "for n-estimator:" + \
str(estimators[RandomForest.index(-min(np.absolute(RandomForest)))]))

plt.figure(figsize=(8,8))
plt.plot(estimators,np.absolute(Bagging),label="BaggingRegressor",c="r")
plt.plot(estimators,np.absolute(Adaboost),label="AdaBoostRegressor",c="b")
plt.plot(estimators,np.absolute(ExtraTree),label="ExtraTreesRegressor",c="g")
plt.plot(estimators,np.absolute(RandomForest),label="RandomForestRegressor",c="y")
plt.xlabel("estimator")
plt.ylabel("mse")
plt.legend(loc=2)
plt.show()
```

```python
# We 'll select RandomForest
# let's look into more detail 

from sklearn.ensemble import  RandomForestRegressor
from sklearn.model_selection import cross_val_score, cross_val_predict

estimators=range(36,47)

def get_mean_mse(model,n):
    ensemble=model(n_estimators=n)
    score=cross_val_score(ensemble,X_train,y_train, cv=5,scoring="neg_mean_squared_error")
    return np.mean(score)


RandomForest=[]


for i in estimators:
    RandomForest.append(get_mean_mse(RandomForestRegressor,i))
    
plt.figure(figsize=(8,8))
plt.plot(estimators,np.absolute(RandomForest),label="RandomForestRegressor",c="b")
plt.xlabel("estimator")
plt.ylabel("mse")
plt.legend(loc=2)
plt.show()
```

```python
#let's use n_estimator = 40.

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

rf = RandomForestRegressor(n_estimators=40)

model2=rf.fit(X_train,y_train)
predictor=model2.predict(X_test)

print " mse:", mean_squared_error(y_test,predictor)
print " r2 score:", r2_score(y_test,predictor)

plt.scatter(y_test, predictor)
plt.ylabel("Predictions")
plt.xlabel("True Values")
plt.plot([0,10],[0,10],c="r",linestyle="--")
plt.xlim((0,10))
plt.ylim((0,10))
plt.show()

```

```python
features=pd.DataFrame(model2.feature_importances_,index=Xt.columns,columns=["Coeff"])
features.sort_values(by="Coeff",inplace=True,ascending=False)
features.head(10)
```
