# Refining the financial and demographic data for each restaurant

The idea in this notebook is to refine the model by optimising Demographic and Financial features. Indeed, each restaurant get the financial and demographic data of the district it belongs even though it should be proportional to the distance to each district.

```python
df_zomato=pd.read_csv("../../assets/Zomato/train_test_zomato.csv")
del df_zomato["Unnamed: 0"]
target=pd.read_csv("../../assets/Zomato/target_zomato.csv")
del target["Unnamed: 0"]
pd.set_option("display.max_columns",200)
```

```python
# This is representation of the distribution of restaurants across lONDON in Zomato

#first encoding the district
df_zomato["District_encode"]=LabelEncoder().fit_transform(df_zomato["District"])

plt.scatter(df_zomato["longitude"],df_zomato["latitude"],c=df_zomato["District_encode"],cmap='rainbow')
plt.ylabel("Latitude")
plt.xlabel("Longitude")
plt.show()
```

```python
# finding the center of each district.

dic_centroid={}

df_zomato["lat_centroid"]=np.nan
df_zomato["long_centroid"]=np.nan

for dist_code in df_zomato["District_encode"].unique():
    lat=np.mean(df_zomato.loc[df_zomato["District_encode"]==dist_code,"latitude"])
    df_zomato.loc[df_zomato["District_encode"]==dist_code,"lat_centroid"]=lat
    longi=np.mean(df_zomato.loc[df_zomato["District_encode"]==dist_code,"longitude"])
    df_zomato.loc[df_zomato["District_encode"]==dist_code,"long_centroid"]=longi
    dic_centroid[dist_code]=(longi,lat)

plt.scatter(df_zomato["long_centroid"],df_zomato["lat_centroid"],c=df_zomato["District_encode"],cmap='rainbow')
plt.title("Center of each district")
plt.ylabel("Latitude")
plt.xlabel("Longitude")
plt.show()
```

```python
# Linking the centroid coordinate to the financial data.

# loading the data
engine=create_engine('postgresql://localhost:5432/capstone')
conn=engine.connect()
```

```python
Dem_Fin=pd.read_sql('select * from "Dem_Fin_1"',con=conn)
Dem_Fin.head()
```

```python
group_code=pd.pivot_table(df_zomato,index=["District"],values=["long_centroid","lat_centroid"],aggfunc=np.mean)
group_code["District"]=group_code.index
```

```python
Dem_Fin=pd.merge(Dem_Fin,group_code,how='left',left_on="Area name",right_on="District")
del Dem_Fin["District"]
del Dem_Fin["index"]

#save in database

Dem_Fin.to_sql("Dem_Fin_2_bis",con=conn,if_exists="replace")
Dem_Fin.head()
```

Now we have the center of each disrict, the next step is to weigth financial and demographic data for each restaurant according to the fourth nearest District centroids.



```python
# calculating distance matrice between each centroid and each restaurant
# calculating the distance matrix
# The below function calculates the distance between any points on the Earth's surface 
# specified by their longitude and latitude in degrees.

from math import radians, cos, sin, asin, sqrt
def haversine(lon1,lat1, lon2,lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2.)**2 + cos(lat1) * cos(lat2) * sin(dlon/2.)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers.
    return c * r
```

```python
matrix_dist=df_zomato[["id","latitude","longitude"]]

for key, val in dic_centroid.items():
    my_metric = matrix_dist.apply(lambda x: haversine(x["longitude"],x["latitude"],val[0],val[1]),axis=1)
    matrix_dist[key]=my_metric
```

```python
#we now have the distance to each centroid
matrix_dist.head()
```

```python
# selecting the 4th closiest centroids and its weight.
def select_4_closier(row):
    listy_dist=row.tolist()
    mini_list=[]
    for i in range(4):
        minimum= mini_list.append(listy_dist.pop(listy_dist.index(min(listy_dist))))
    min4=mini_list[-1]
    mini_list.sort()
    for j in range(len(row)):
        if row[j]-min4>0:
            row[j]=0
        else: # the smaller minimum take the higher rate, we inverse
            row[j]=mini_list[3-mini_list.index(row[j])]
    row=row/sum(row)
    return row

matrix_dist.iloc[:,3:]=matrix_dist.iloc[:,3:].apply(select_4_closier,axis=1)

group=pd.pivot_table(df_zomato,index=["District"],values=["District_encode"],aggfunc=np.mean)
zone=group.index.tolist()

matrix_dist.columns=["id","latitude","longitude"]+zone

matrix_dist.head()
```

```python
# apply the weighted financial data to df_zomato.

del df_zomato['lat_centroid']
del df_zomato['long_centroid']

Dem_Fin.sort_values(by="Area name",inplace=True)
Dem_Fin.reset_index(inplace=True,drop=True)

col_finance=['GLA Population Estimate 2015',
 'GLA Household Estimate 2015',
 'Inland Area -hec',
 'Population density - Hec-2015',
 'Average Age, 2015',
 'Employment rate - 2014',
 'Unemployment rate-2014',
 'Gross Annual Pay-2014',
 'Modelled Household median income estimates 2012/13',
 'Number of jobs by workplace-2013',
 'Jobs Density, 2013',
 'Two-year business survival rates-2011/2015',
 'Median House Price, 2014',
 'Average Public Transport Accessibility score, 2014',
 'Perc. People aged 17+ with diabetes ']


for col in col_finance:
    df_zomato[col]=matrix_dist.iloc[:,3:].dot(Dem_Fin[col].as_matrix())
```
