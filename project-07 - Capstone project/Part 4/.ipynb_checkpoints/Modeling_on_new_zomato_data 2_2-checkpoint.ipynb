# Calculating the new Target


## Reminder of Restaurant success categorisation

We now have to categorise the health of a business so that our model predict the best location at the moment for a type of food.
How to categorise?
Well, we could look at the profit margin: Profit = Revenue - Cost.
Of course, I don’t have all parameters to calculate the exact profit but as we only want to compare restaurants to each other, we can consider only the parameter relative to each restaurant. Which means, electricity bill, wages, etc are actually constant through all restaurant but other such as revenue , cost of ingredients, rental expenses

We are the reducing the problem to:

**Profit_reduced ≈ Revenue - [ Cost_ing + Cost_rent]**


__Revenue:__

Keeping with consideration because, as a recall we want to categorise restaurant, the revenue can be considered proportional to the number of vote. Indeed, the number of votes being proportional to the attendance level.
Although we are not taking into consideration the year of service, we are fitting a model on our actual static data. Thus, after the model being shaped, it could be scaled to the weekly/monthly evolution of the number of votes

**Revenue ≈ K1 \* (Number of votes) \* cost_for_two **

_On average 1 vote for 2 persons_

__Cost ingredients:__

Ingredient will be chosen in function of the cuisine and quality you want to propose. Although, it is common that the ingredient should represent between 25% to 45% of the price propose to the customer and the quality is linked to the price range category. In other words, with the data we have, we can rewrite the Cost_ing as such:

**Cost_ing ≈ F(cuisine, quality) ≈ F(cuisine, cost_for_2, price_range) ≈  Average(cuisine, 35%*cost_for_two ,price_range)*(Number of votes) **

__Cost of the place:__

I am here presuming that the surface area for each restaurant are more or less the same as well as all of them having rental,mortgage,.., expenses. Of course, this is not totally the case but I think close to the reality.

As I have got the median house price in London, accommodation expenses will be proportional to the rental expenses from where the district has the average minimum house price multiply by a factor. The factor being the median house price of the District divided by a minimum median house price.

**location\_coeff ≈ K2* (Median_house_price_district / Minimum(median_house_price_all_district) )**


__In a nutshell:__


**Profit_reduced ≈ K1* (Number of votes)*cost_for_two**
	**- Average(cuisine, 35%\*cost_for_two,price_range)\*(Number of votes) ** 
	**- K2* (Median_house_price_district / Minimum(median_house_price_all_district))**
    
The aim of this formula is not to get the exact profit, because probably it contains some inconsistencies. Moreover, as the cost of the place is in a totally different scale, it makes no sense to retrieve it. 

That why, we will use the coefficient Median_house_price_district / Minimum(median_house_price_all_district) as a tuner. Indeed, if two restaurants have the same cuisine and selling price, to equalise their profit, or they have the same accommodation charges or the first restaurant have a number of customer proportional to rent of the other.

We would then get the following formula:

**Profit_reduced ≈ (K1* (Number of votes)*cost_for_two**
	**- Average(cuisine, 35%\*cost_for_two,price_range)\*(Number of votes))/Location\_coeff ** 
    
Location\_coeff=Median\_house\_price\_district / Minimum(median\_house\_price\_all\_district) 

Consequently, after scaling each feature of the formula, we would be able to find cluster of restaurant taking into account those 3 features (revenue and costs). The cluster created will permit to categorise the restaurant.


```python
# Only the new location coeeff variable will affect the profit tuned
# consequently we have the equality New_Profit_tuned = Old_Porfit_Tuned*OLD_location_coeff/New_location_coeff
# New_Profit_tuned=OLD_RESULT/New_location_coeff

OLD_RESULT= df_zomato["profit_tuned"]*df_zomato["location_coeff"]


df_zomato["location_coeff"]=df_zomato["Median House Price, 2014"]/min(df_zomato["Median House Price, 2014"])
df_zomato["profit_tuned"]=OLD_RESULT/df_zomato["location_coeff"]
df_zomato.head()
```

# Grouping the profit_tuned in 5 category

As done in part 3, I am now creating 5 categories of likelihood of success with their own label:

- Totaly likely : 5
- Very Likely: 4
- Likely: 3
- Few chance: 2
- Very few chance: 1

Indeed, in a big city as London I am considering that there success of restaurants follow a +/- a normal distribution, having restaurant with great success, others without and the main main part with a limited success. 

The distribution will be:
- cat 5 : 5%
- cat 4 : 25%
- cat 3 : 40%
- cat 2 : 25%
- cat 1 : 5%

```python
print "-Not very likely chances of success for profit_tuned below %.2f"%df_zomato["profit_tuned"].quantile(0.20)
print "-Few chances of success for profit_tuned between %.2f and %.2f"%\
(df_zomato["profit_tuned"].quantile(0.05),df_zomato["profit_tuned"].quantile(0.3))
print "-Likely to get success for profit_tuned between %.2f and %.2f"%\
(df_zomato["profit_tuned"].quantile(0.3),df_zomato["profit_tuned"].quantile(0.70))
print "-Very Likely to get success for profit_tuned between %.2f and %.2f"%\
(df_zomato["profit_tuned"].quantile(0.70),df_zomato["profit_tuned"].quantile(0.95))
print "-Totaly Likely to get success for profit_tuned above %.2f"%\
(df_zomato["profit_tuned"].quantile(0.95))

def get_category(row):
    category=[1,2,3,4,5,6]
    threshold=[df_zomato["profit_tuned"].quantile(0.05),\
              df_zomato["profit_tuned"].quantile(0.3),\
              df_zomato["profit_tuned"].quantile(0.70),\
              df_zomato["profit_tuned"].quantile(0.95),max(df_zomato["profit_tuned"])]
    check= row <= threshold[0]
    index=0
    while check==False:
        index+=1
        check= row<=threshold[index]
    
    return category[index]
    
 
df_zomato["target"]=df_zomato["profit_tuned"].apply(get_category)
```

```python
# let's select all features to perform a PCA except group_cuisine dummies that we will join after.

col_pca=[\
 'price_range',\
 'latitude',\
 'longitude',\
 'area_cover',"cluster", \
 'restaurant_nearby',\
 'GLA Population Estimate 2015',\
 'GLA Household Estimate 2015',\
 'Inland Area -hec',\
 'Population density - Hec-2015',\
 'Average Age, 2015',\
 'Employment rate - 2014',\
 'Unemployment rate-2014',\
 'Gross Annual Pay-2014',\
 'Modelled Household median income estimates 2012/13',\
 'Number of jobs by workplace-2013',\
 'Jobs Density, 2013',\
 'Two-year business survival rates-2011/2015',\
 'Median House Price, 2014',\
 'Average Public Transport Accessibility score, 2014',\
 'Perc. People aged 17+ with diabetes ']

X_std=df_zomato[col_pca]
X_std = StandardScaler().fit_transform(X_std)
pca=PCA()
X_pca = pca.fit_transform(X_std)

exp_var=pca.explained_variance_ #explained variance of the principal components
exp_var_rat=pca.explained_variance_ratio_ #ratio of total variance

cum_var=np.cumsum(exp_var)/sum(exp_var)

plt.plot(range(1, len(cum_var) + 1), cum_var)
plt.title("Cumulative proportion of Variance Explained")
plt.ylabel("Variance Explained")
plt.xlabel("Number of Principal Components Included")
plt.xlim(1,21)
plt.show()
```

There is not a clear elbow (point non-differentiable) exept for 2 components, the 2 components representing 60% of the variance explained. Also we can notice a non-differentiable point for 8 principal components where the 8 components explain more than 90% of the variance.
I am then selecting the 8 first principal components and will appply the models on it.

```python
#As we did in previous notebook, I am drawing a fonction to plot confusion matrix

import itertools

def plot_confusion_matrix(cm, classes, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    return
```

# Selecting the output based on its frequency appearance

With no interest in performing principal component analysis, I am testing the effect on the normal set on comparing the output from all combinations of each classifier output

```python
# I am now selecting the ouptut in function on the average probability.

col_prediction= [predicion_proba_gradboost,predicion_proba_logit,predicion_proba_svc,predicion_proba_dtc,predicion_proba_xgb]
col_prediction_name= ["predicion_proba_gradboost","predicion_proba_logit","predicion_proba_svc","predicion_proba_dtc","predicion_proba_xgb"]



def select_output(row):
    ind=1
    output_prob=row[1]
    for i in range(2,6):
        if row[i]>=output_prob:
            ind=i
            output_prob=row[i]
    return ind

for i in range(2,6):
    # creating all combination possible of i models
    comb=itertools.combinations(col_prediction, i)
    comb_name=list(itertools.combinations(col_prediction_name, i)) # getting the name of those combination.
    
    indice=0
    for item in comb:
        print "COMBINATION:", comb_name[indice]
        average=sum(item)/len(item)
        
        output_proba=pd.DataFrame(average,columns=[1,2,3,4,5])


        output_proba["cat"]=output_proba.apply(select_output,axis=1)
        test_target=y_test_main.copy()
        test_target.reset_index(inplace=True, drop=True)
        output_proba["target"]=test_target

        print "Output selection based on average proba:"
        print metrics.confusion_matrix(y_test_main,output_proba['cat'])

        print "Accuracy:",metrics.accuracy_score(y_test_main,output_proba['cat'])
        print "Classification report:\n", metrics.classification_report(y_test_main,output_proba['cat'])
        print "------\n"
        indice=indice+1
```

```python
# I am now selecting the ouptut in function on the frequency appearance for the output 

col_prediction= [prediction_gradboost,prediction_logit,prediction_svc,prediction_dtc,prediction_xgb]
col_prediction_name= ["predicion_gradboost","predicion_ogit","predicion_svc","predicion_dtc","predicion_xgb"]


for i in range(2,6):
    # creating all combination possible of i models
    comb=itertools.combinations(col_prediction, i)
    comb_name=list(itertools.combinations(col_prediction_name, i)) # getting the name of those combination.
    
    indice=0
    for item in comb:
        print "COMBINATION:", comb_name[indice]
        print len(comb_name)
        
        array_final=np.reshape(item[0],(len(item[0]),1))
        for k in range(1,len(item)):
            array_final=np.concatenate((array_final,np.reshape(item[k],(len(item[k]),1))),axis=1)
        output_frequency=pd.DataFrame(array_final)
        output_frequency.columns=[comb_name[indice][j] for j in range(output_frequency.shape[1])]
        
        def select_frequency(row): 
            dic_cat={}
            for i in range(output_frequency.shape[1]):
                if row[i] in dic_cat:
                    dic_cat[row[i]]=dic_cat[row[i]]+1
                else:
                    dic_cat[row[i]]=1
            max_cat=max(dic_cat.values())
            # if equality return the outliers
            key_list=dic_cat.keys()
            value_list=dic_cat.values()
            count=0
            maximum=[]
            for i in range(len(key_list)):
                if value_list[i]==max_cat:
                    count+=1
                    maximum.append(key_list[i])
            if count==1:
                result= dic_cat.keys()[dic_cat.values().index(max_cat)]
            else:
                outliers=[1,5]
                if outliers[1] in maximum:
                    result=outliers[1]
                elif outliers[0] in maximum: #conservative model
                    result=outliers[1]
                else:
                    result=dic_cat.keys()[dic_cat.values().index(max_cat)]
            
            return result
        
        test_target=y_test_main.copy()
        test_target.reset_index(inplace=True, drop=True)
        output_frequency["cat"]=output_frequency.apply(select_frequency,axis=1)
        output_frequency["target"]=test_target

        print "Output selection based on frequency output appearance:"
        print metrics.confusion_matrix(y_test_main,output_frequency['cat'])

        print "Accuracy:",metrics.accuracy_score(y_test_main,output_frequency['cat'])
        print "Classification report:\n", metrics.classification_report(y_test_main,output_frequency['cat'])
        print "------\n"
        indice=indice+1
    
```

# Calculating the degree of confidence to be inside an interval of success

Various model gave us an accuracy over 50 and a precision over 45. Also, it seems that even if the model is far for being accurate, it manage to give an answer close to right one. It seems an interval of confidence of cat +/- 1 can well describe the situation of the distribution.
For example, a model where cat 4 is predicted with an accuracy of 50% but if almost all predicted cat 4 are in an interval cat 4- cat 5, then the model predicted that all those restaurant has indeed a chance of success.
I also remind that, even if the location is the great part of the success, other factors such as quality of the food, marketing, etc have also an important effect it.

In the next cells, I am listing 10 models that had a great result in term of Accuracy, precision.
The I am calculating which one degree of confidence that the result is being in a range of +/- 1 category.
